```python
#Karar yapıları:Programın çalışma akışını belirli koşullara göre yöneten,hangi adımın çalışacağına karar veren yapılardır
#İf Deyimi: Bir koşul doğruysa belirli kodların çalışmasını sağlayan karar yapısıdır.
x=int(input("Sayı girin:"))
if x>5:
    print("x,5'ten büyüktür.")#Koşul doğruysa çalışacak kod bloğu
print("Program bitti.")
```

```python
#İf-Else yapısı
yas=int(input("Sayı girin:"))
if yas>=18:
    print("Reşitsiniz")
else:#Önceki kod bloğu çalışmazsa çalışan kod bloğu fffffweıgfervgrehjsvbfhvf
    print("Reşit değilsiniz(18 yaşından küçüksünüz.)")
print("Program bitti.")
```

```python
#İf-Elif-Else deyimi
sayi=int(input("Sayı girin:"))
if sayi>0:#Koşul1 doğru ise çalışacak kod bloğu
    print(f"Girilen Sayı {sayi:.2f}, pozitiftir")#Tab içeri girmezsek hata alırız.
elif sayi<0:#Koşul1 yanlış,Koşul2 doğru ise çalışacak kod bloğu//istediğimiz sayıda elif bloğu ekleyebiliriz.
    print(f"Girilen Sayı {sayi:.2f}, negatiftir")
else:#Hiçbir kod bloğu doğru değilse çalışacak kod bloğu
    print(f"Girilen sayı {sayi:.2f}, sıfırdır")
```
